```python
c.execute('drop table if exists MenuPan')
c.execute('create table MenuPan(menuName char(50), price int)')
c.execute('insert into MenuPan values("김밥", 3000)')
c.execute('insert into MenuPan values("순두부", 6000)')
c.execute('insert into MenuPan values("육개장", 7500)')
c.execute('insert into MenuPan values("칼국수", 5500)')
c.execute('insert into MenuPan values("짜장밥", 5500)')
c.execute('insert into MenuPan values("김치찌개",8000)')
c.execute('insert into MenuPan(price, menuName) values(800,"aaa")')
c.execute('insert into MenuPan(price, menuName) values(666,"bbb")')

conn.commit()
```

```python
foodName = input('food name : ')
# 입력받을 것을 숫자처리함
foodPrice = int(input('food price : '))

# 변수 처리--> ?
# MenuPan(price, menuName) 이름 순서를 따로 쓰는 이유는
# insert할 때 순서가 바뀔 수 있기 때문에 명시적으로 써주는 것이 좋다.
# 안쓸 경우 원래 테이블의 순서대로 입력해야 한다.
c.execute('insert into MenuPan(price, menuName) values(?, ?)', (foodPrice, foodName))
conn.commit()

c.execute('select * from menuPan')
rows = c.fetchall()

for row in rows:
  print("{} {}".format(row[0], row[1]))
```

```python
def menuPanSelect():
    c.execute('select * from menuPan')
    rows = c.fetchall()

    for row in rows:
        print("{} {}".format(row[0], row[1]))
```

```python
menuPanSelect()
```

```python
# update
foodName = input('food name : ')
chagePrice = int(input('chage food price : '))

c.execute('update menupan set price = ? where menuName= ?', [chagePrice, foodName])
conn.commit()
print('\n\n')
menuPanSelect()
```

```python
# 메뉴 이름 변경하기
menuPanSelect()

foodName = input('food name : ')
# changePrice = int(input('변경할 가격 : '))
changeFoodName = input('바꿀 이름 : ')
c.execute('update menupan set menuname = ? where menuname = ?', [changeFoodName, foodName])
conn.commit()
print('\n\n')
menuPanSelect()
```

```python
# 레코드 삭제 : delete
# delete from 테이블명 # 전체 레코드 삭제
# delete from menupan where 조건 # 전체 레코드 삭제
```

```python
menuPanSelect()
# delete from menupan where menuName='aaa'
c.execute('delete from menupan where menuName = "bbb"')
print("---------------------------------")
menuPanSelect()
```

```python
# 변수처리하여 입력받아 레코드 지우기
menuPanSelect()
# delete from menupan where menuName='aaa'
temp = input('삭제할 메뉴 이름 : ')
c.execute('delete from menupan where menuName = ?', [temp])
print("---------------------------------")
menuPanSelect()
```
