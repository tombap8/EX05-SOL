```python
# name, kor, eng, com 점수 입력 받아서
# 총점, 평균, 학점 계산하는 프로그램 작성

# ***** 도연님의 성적표 ******
# 이름 : 도연
# 국어 : 90점, 영어 : 85점, 컴퓨터 : 95점
# 총점 : 270점, 평균 : 90점, 학점 : A학점
```

```python
name = input("이름을 입력하세요: ")
kor = int(input("국어 점수를 입력하세요: "))
eng = int(input("영어 점수를 입력하세요: "))
com = int(input("컴퓨터 점수를 입력하세요: "))

total = kor + eng + com
avg = total / 3

if avg >= 90:
    grade = 'A'
elif avg >= 80:
    grade = 'B'
elif avg >= 70:
    grade = 'C'
elif avg >= 60:
    grade = 'D'
else:
    grade = 'F'

avg_display = int(avg) if avg.is_integer() else round(avg, 2)

print(f"\n***** 출력 예시 {name}님의 성적표 ******")
print(f"이름 : {name}")
print(f"국어 : {kor}점, 영어 : {eng}점, 컴퓨터 : {com}점")
print(f"총점 : {total}점, 평균 : {avg_display}점, 학점 : {grade}학점")
```
